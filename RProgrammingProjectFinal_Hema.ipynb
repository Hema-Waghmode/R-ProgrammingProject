```r
# Load the required libraries
install.packages("rvest")
library(rvest)
library(stringr)
library(rvest)
library(xml2)


url <- "https://www.goibibo.com/flights/?utm_source=google&utm_medium=cpc&utm_campaign=DF-Brand-EM&utm_adgroup=Only+Goibibo&utm_term=%21SEM%21DF%21G%21Brand%21RSA%21108599293%216504095653%21602838584772%21e%21goibibo%21c%21&gclid=CjwKCAjw_YShBhAiEiwAMomsEA_BLUAYmGqIb4nMXSrYS8OCLAH0uidaBKeVYmEkrAi6fcyaKpVwoRoCLLsQAvD_BwE"


# Send a GET request to the URL and parse the HTML response
page <- read_html(url)

# Extract the product title, price, and rating for the first 10 search results
mainMenus <- page %>% 
  html_nodes(".sticky-gutter") %>%
  html_text()

LmainMenus <- page %>%
  html_nodes(".nav-link") %>%
  html_text()

footer <- page %>%
  html_nodes(".happy-gi-footer") %>%
  html_text()

popularRoutes <- page %>%
  html_nodes(".sc-iuzt5e-3") %>%
  html_text()


faqs <- page %>% 
  html_nodes(".sc-vqihtk-2") %>%
 
  html_text("aria-label")
 # html_attr("aria-label")


# Print the scraped product information
cat("Main Menus:\n", mainMenus[1:10], "\n")
cat("Left Side Main Menus:\n", LmainMenus[1:10], "\n")
cat("Footer:\n", footer[1:10], "\n")
cat("Frequently Asked Questions:\n", faqs[1:10], "\n")
cat("PoPular Routes:\n", popularRoutes[1:10], "\n")






```

```r
#Dealing with NA values
#Since, the distribution of Income and Age is left-skewed. We will impute median values

#The na. omit R function removes all incomplete cases of a data object
# (typically of a data frame, matrix or vector).
median(na.omit((bike_buyers$Income)))
median(na.omit((bike_buyers$Age)))
```

```r
bike_buyers_clean <- bike_buyers
colSums(is.na(bike_buyers_clean))
```

```r
# Income replaced with Median
bike_buyers_clean$Income[is.na(bike_buyers_clean$Income)] <- 
  median(na.omit((bike_buyers$Income)))

# Age replaced with Median
bike_buyers_clean$Age[is.na(bike_buyers_clean$Age)] <- 
  median(na.omit((bike_buyers$Age)))

colSums(is.na(bike_buyers_clean))
```

```r
#Save clean dataframe¶
write.csv(bike_buyers_clean,"bike_buyers_clean.csv", quote = FALSE, row.names = TRUE)
```

```r
#Bar Plot
#A bar plot is a plot that presents categorical data with rectangular bars with lengths proportional to the 
#values that they represent.
counts <- table(bike_buyers$Cars, bike_buyers$Gender)
barplot(counts, main = '',
        xlab="Gender",
        legend = rownames(counts),col="green",border = "red")
```

```r
#Scatter Plot

#A scatterplot is a type of data display that shows the relationship between two numerical variables.
#Scatter plots show how much one variable is affected by another. 
plot(bike_buyers$Income,pch=1)
```

```r
#Histogram
#A histogram is a graphical display of data using bars of different heights. In a histogram, each bar groups numbers into ranges.
ggplot(bike_buyers, aes(x = Age)) +
  geom_histogram()
```

```r
#Density Plot
#A density plot is a representation of the distribution of a numeric variable.
plot(density(bike_buyers$Income), main='Income Density Spread')
```
